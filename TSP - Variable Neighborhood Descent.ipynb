Ciò che vogliamo fare è partire da un percorso, calcolarne il fitness, cercare altri percorsi vicini, calcolarne il loro fitness e prendere il migliore, a questo punto mantengo un posizione e cerco altri vicini e ripeto l'operazione.
Esempio: Percorso di partenza = 10,20,30 = 60fitness, trovo un percorso 10,20,20 = 50fitness, a questo punto fisso un tratto e cerco i suoi vicini, quindi avrò per esempio 10,15,60 o 10,15,15 = 40fitness, a questo punto prendo questo e fisso anche la seconda posizione, avendo quindi fissati 10,15, e cercando l'ultimo valore ecc..
In questo modo ogni volta che trovo una soluzione migliorativa esploro i dintorni.

```python
import numpy as np
import random
```

```python
#Come fitness utilizzo la somma delle lunghezze del percorso
def fitness(solution):
    fitness = solution.sum()
    return fitness
```

```python
#Dato un array, questo ci pesca dentro un numero casuale, in questo caso si passa anche 
#la posizione iniziale, che non deve essere ripescata e un array contenente le posizioni già uscite, quindi i punti
#già visitati
def random_number(array,tmp_position,starting_position):
    while True:
        indice, valore = random.choice(list(enumerate(array)))
        if valore != 0 and indice not in tmp_position and indice != starting_position:
            return indice, valore
```

```python
def vnd(matrix,starting_solution,starting_position,k_max,dimension_matrix):
    
    #Lista che tiene conto di tutti i fitness trovati così da fare un plot
    total_fitness = np.array([])
    
    #Passiamo una soluzione iniziale casuale da valutare
    print("Soluzione INIZIALE: " + str(starting_solution))
    print("Fitness: " + str(fitness(starting_solution)))
    print()
    solution = starting_solution.copy()
    path_solution = np.full(dimension_matrix, -1, dtype=int)
    
    k = 0 #Indica il numero di vicini che vogliamo generare prima di fermarci
    position = starting_position #Indica la posizione da cui vogliamo partire nel grafo
    
    #Quando generiamo i vicini, questo good_solution ci fa da indice per tener conto dei percorsi che non vogliamo
    #sostituire, infatti se io trovo una soluzione migliorativa [10,20,30,40], voglio generare i vicini tenendo il 10
    #e quindi il mio good_solution varrà 1, in modo da modificare l'array dal valore 20 in poi, non toccando il 10.
    good_solution = 0 
    
    index_tmp_solution = 0 #Un indicice che tiene conto della posizione da riempire da mettere in tmp_solution e tmp_position
    tmp_solution = np.zeros(dimension_matrix, dtype=int) #L'array con una soluzione temporanea, inizializzato come [0,0,0,0]
    tmp_position = np.full(dimension_matrix, -1, dtype=int)#L'array che indica verso che punto ci stiamo muovendo (inizializzato a -1)
    
    #tmp_position e tmp_solution sono connessi, quindi se ho tmp_position = [10,20,30,40] allora tmp_solution = [0,1,2,3]
    #significa che il percorso che costa 10 va verso il punto 0, poi da 0 a 1 il percorso costa 20, ecc...
    
    #Itero finchè non ho generato tutti i vicini
    while k < k_max:
        #Vogliamo riempire l'array con tutti i percorsi partendo da un punto e tornando alle stesso visitando tutto
        while index_tmp_solution < dimension_matrix:
            array = matrix[position] #prelevo le strade possibile di un punto, quindi una riga
            #Se dobbiamo prendere l'ultimo tratta, questa è sicuramente quella che ci riporta all'inizio 
            if index_tmp_solution == dimension_matrix-1: 
                index, path = starting_position, array[starting_position]
            #Se invece non è l'ultima tratta, prendo una strada casuale partendo dal punto in cui mi trovo
            else:
                index, path = random_number(array,tmp_position,starting_position)
                
            tmp_solution[index_tmp_solution] = path #salvo il valore prese
            tmp_position[index_tmp_solution] = index #salvo la posizione in cui mi trovo
            index_tmp_solution = index_tmp_solution + 1
            position = index #setto il nuovo punto che ho raggiunto (in mo do da predere successivamente le sue strade)
        
        #Se ancora non c'è stata nessuna soluzione migliorativa, semplicemente resetto l'index, altrimenti 
        #resetto l'index ma non parto da 0, ma parto da un'altro valore,per esempio 1, 
        #questo poichè voglio generare dei vicini partendo dalla soluzione migliorativa (quindi mantengo una strada)
        #e cambio le altre
        index_tmp_solution = good_solution
        
        #Mi salvo il valore di fitness ottenuto
        total_fitness = np.append(total_fitness, fitness(tmp_solution))
        
        print("Soluzione temporanea: " + str(tmp_solution))
        print("Fitness: " + str(fitness(tmp_solution)))
        print("Percorso: " + str(tmp_position))
        #Se la soluzione trovata è migliorativa
        if fitness(solution) > fitness(tmp_solution):
            print("Soluzione temporanea migliorativa")
            #mi salvo la soluzione migliorativa e la strada
            solution = tmp_solution.copy()
            path_solution = tmp_position.copy()
            #imposto una nuova posizione di partenza (salvando le precedenti o la precedente)
            position = tmp_position[good_solution]
            index_tmp_solution = good_solution + 1
            good_solution = good_solution + 1
            k = 0 #resetto i tentativi
            #se non abbiamo già visitato "abbastanza" vicini (quindi abbiamo fissato meno di 4 percorsi),
            #resetto i percorsi e le loro rispettive posizioni, mantenendo quelli che ho fissato
            if good_solution < dimension_matrix:
                for i in range(index_tmp_solution, len(tmp_solution)):
                    tmp_solution[i] = 0
                    tmp_position[i] = -1
                print("Genero i vicini partendo da: " + str(tmp_solution))
                print("Posizioni salvate: " + str(tmp_position))
        else:
            #Se la soluzione non è migliorativa, aggiorno i tentativi
            k = k + 1
            #Sistemo la nuova posizione da cui partire
            if(good_solution != 0):
                position = tmp_position[good_solution-1]
            else:
                position = starting_position
            #resetto i percorsi mantenendo quelli già fissati
            if good_solution < dimension_matrix:
                for i in range(good_solution, len(tmp_solution)):
                    tmp_solution[i] = 0
                    tmp_position[i] = -1
        print()


    print("BEST solution: " + str(solution))
    print("BEST path (Start point = " + str(starting_position) + "): " + str(path_solution))
    print("Fitness: " + str(fitness(solution)))
    return total_fitness
    
        
```
